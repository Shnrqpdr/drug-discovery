# Import modules

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import utils
sns.set()
```

# Read the dataframe

```python
url_dados = 'https://github.com/alura-cursos/imersaodados3/blob/main/dados/dados_experimentos.zip?raw=true'

dados = pd.read_csv(url_dados, compression='zip')

dados
```

```python
dados.rename(columns={'droga': 'composto'}, inplace=True)

dados
```

# First analysis

A principal ideia dessa primeira análise de dados é entender a proporçao entre as moléculas que foram tratadas com drogas e as moléculas que foram tratadas com controle.

```python
comDroga = dados['tratamento'].value_counts()[0]
comDrogaPorcentagem = dados['tratamento'].value_counts('percent')[0]*100.0

comControle = dados['tratamento'].value_counts()[1]
comControlePorcentagem = dados['tratamento'].value_counts('percent')[1]*100.0
    
barWidth = 0.40

fig, axes = plt.subplots(1, 3, figsize=(24,6))

axes[0].bar('Com droga', comDroga, width=barWidth, label="Quant de moléculas estudadas com droga")
axes[0].bar('Com controle', comControle, width=barWidth, label="Quant de moléculas estudadas com controle")
axes[0].set_ylabel("Quantidade")
axes[0].set_title(f'Comparação da quantidade de moléculas estudadas')
axes[0].grid(axis='y', linestyle='-', linewidth=0.5)
axes[0].legend(loc='upper right')

axes[1].bar('Com droga', comDrogaPorcentagem, width=barWidth, label="Quant de moléculas estudadas com droga")
axes[1].bar('Com controle', comControlePorcentagem, width=barWidth, label="Quant de moléculas estudadas com controle")
axes[1].set_ylabel("Quantidade em porcentagem (%)")
axes[1].set_title(f'Comparação da quantidade de moléculas estudadas')
axes[1].grid(axis='y', linestyle='-', linewidth=0.5)
axes[1].legend(loc='upper right')

axes[2].pie(dados['tratamento'].value_counts(), labels=['Com droga', 'Com controle'], autopct='%1.1f%%', startangle=90)
axes[2].axis('equal')

plt.show()
```

```python
compostosMaisUtilizados = dados['composto'].value_counts()[0:5]
codigoCompostosMaisUtilizados = compostosMaisUtilizados.index

dadosFiltrados = dados.query('composto in @codigoCompostosMaisUtilizados')

dadosFiltrados

plt.figure(figsize=(10,8))
ax = sns.countplot(x = 'composto', data=dadosFiltrados, order=dadosFiltrados['composto'].value_counts().index)

ax.set_title('5 Compostos mais utilizados')
ax.set_xlabel('Compostos')
ax.set_ylabel('Quantidade')

plt.show()
```
